```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('astronum_speedups_2017.csv', delimiter=',')[5:9]
```

```python
data
```

```python
d2 = []
for r in data:
    d2.append(r[1:])
```

```python
d2
```

```python
n = [r[0] for r in d2]
aprox13 = [r[1] for r in d2]
urca = [r[2] for r in d2]
ignsimple = [r[3] for r in d2]
```

```python
plt.clf()
plt.plot(n, ignsimple, color='green', marker='o', label='3-Isotope Carbon Ignition')
plt.plot(n, urca, color='blue', marker='o', label='9-Isotope A=23 Urca Process')
plt.plot(n, aprox13, color='red', marker='o', label='13-Isotope Alpha Chain')
plt.ylim([1,40])
plt.yticks([1, 5, 10, 20, 30, 40], fontsize=14)
plt.xticks([16, 32, 64, 128], fontsize=14)
plt.xlabel('Cubic Grid Width', size=14)
plt.ylabel('Speedup: Tesla P100 vs. Serial POWER8', size=14)
lgd = plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.grid('on')
plt.savefig('cuda-vode-speedup.png', dpi=600, bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
